grouping and Aggregation

groupby():Grouping data by one or more columns for aggregation

```python
grouped_df=df.groupby('city')['price'].sum()
grouped_df
```

agg():Appiying aggreation functions(e.g.,sum,mean)to grouped data

```python
agg_df=df.groupby('category').agg({'size': 'mean'})
print (agg_df)
```

count():Counting non-nullvalues in each group.

```python
count_df=df.groupby('category')['size'].count()
count_df
```

Data visualisation using pandas where backend is matplotlib itself

LINE PLOT:--A line plot is useful for visualizing trends over time.

```python
import matplotlib.pyplot as plt
df.plot(x='category',y='price',kind='line',marker='o',figsize=(10,5))
```

BAR PLOT:--A bar plot is used to compare the frequency or count of item across diffrent categories.

```python
df.plot(x='price',y='size',kind='bar',
color='skyblue',figsize=(10,10))
```

HISTOGRAM:--A histogram is used to visualize the distribution of a dataset.

```python
df['room_count'].plot(kind='hist',color='green',edgecolor='black')
```

SCATTER PLOT:-A scatter plot is useful for visualizing the relationship between two variables.

```python
df.plot(x='bathroom_count',y="price",kind='scatter', color='red')
```
